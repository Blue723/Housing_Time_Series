```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.pylab as plb
from statsmodels.tsa.stattools import adfuller

%matplotlib inline
```

```python
df = pd.read_csv('zillow_data.csv')
```

```python
df.info()
```

```python
df_set = df[(df.State == 'TX') & (df.City == 'Houston')]
df_set
```

```python
df_set = df_set.sort_values(by='SizeRank')
df_set
```

```python
df_set = df_set[:15]
df_set
```

```python
df_set.isna().sum()
```

Filtering dataset to the top 15 Size Rank in Houston,TX Harris County. 

```python
df_set.reset_index(drop=True, inplace=True)
```

```python
df_info = df_set.loc[:, ['RegionID', 'RegionName', 'City', 'State', 'Metro', 'CountyName',
       'SizeRank']]
```

```python
final_df = pd.concat([df_info, df_set.loc[:,'2012-01': '2017-12']], axis=1)
```

```python
def get_datetimes(df):
    return pd.to_datetime(df.columns.values[7:], format='%Y-%m')
```

```python
date_time_idx = get_datetimes(final_df)
```

```python
date_time_idx
```

```python
date_time_idx = date_time_idx.strftime('%Y-%m')
```

```python
date_time_idx
```

```python
def melt_data(df):
    melted = pd.melt(df, id_vars=['RegionName', 'RegionID', 'SizeRank', 'City', 'State', 'Metro', 'CountyName'], var_name='time')
    melted['time'] = pd.to_datetime(melted['time'], infer_datetime_format=True)
    melted = melted.dropna(subset=['value'])
    return melted.groupby('time').aggregate({'value':'mean'})
```

```python
zipcode_data = {}
for i in range(0,15):
    zipcode_data[str(final_df.loc[i, 'RegionName'])] = melt_data(pd.DataFrame(final_df.iloc[i:i+1, :], columns=final_df.columns))
```

```python
z = pd.DataFrame()
for key in zipcode_data.keys():
    z = pd.concat([z , zipcode_data[key]], axis=1)
```

```python
zipcode = pd.DataFrame(z.values, columns=zipcode_data.keys())
```

```python
zipcode.info()
zipcode.head()
```

```python
zipcode = zipcode.set_index(date_time_idx)
```

```python
zipcode.head()
```

```python
zipcode.plot();
```

```python
def stationarity(df):
    roll_mean = df.rolling(window=12, center=False).mean()
    roll_std = df.rolling(window=12, center=False).std()
    
    fig = plt.figure(figsize=(12,6))
    plt.plot(df, color='blue',label='Original')
    plt.plot(roll_mean, color='red', label='Rolling Mean')
    plt.plot(roll_std, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()
    return None
```

```python
for c in zipcode.columns:
    stationarity(zipcode[c])
```

```python
# Perform Dickey-Fuller test:
print ('Results of Dickey-Fuller Test: \n')
dftest = adfuller(zipcode['77084'])

# Extract and display test results in a user friendly manner
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
print(dfoutput)
```
